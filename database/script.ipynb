# [dbo].[cant_res_dia]

```sql
USE [hotel]
GO
```

```sql
/****** Object:  UserDefinedFunction [dbo].[cant_res_dia]    Script Date: 20/06/2023 20:01:21 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create function [dbo].[cant_res_dia] (@dia int)
	returns int as 
		begin
			declare @cantidad int = 0
			set @cantidad = (select count (reserva.fecha) from reserva where reserva.fecha = @dia or reserva.fecha = @dia+1 or reserva.fecha = @dia+2)

	return @cantidad
end
GO
```

# [dbo].[evaluacion01]

```sql
/****** Object:  UserDefinedFunction [dbo].[evaluacion01]    Script Date: 20/06/2023 20:01:21 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

--halla el promedio de los precios y si es mayor igual que el gasto return true(hotel de resultado)
create function [dbo].[evaluacion01] (@tipo int, @gasto int ) --alter
	returns int as 
		begin
		declare @suma int = 0
		declare @contador int = 0
		declare @promedio int = 0
		declare @aux bit = 0
			
		while exists (select * from habitacion where tipo <= @tipo)
			begin		
				set	@suma += (select sum(precio) from habitacion where tipo = @tipo)
				set	@contador += (select count(precio) from habitacion where tipo = @tipo)
				set @tipo -= 1			
		   end

		set @promedio =  @suma / @contador

		if (@promedio >= @gasto)
			begin 
				set @aux = 1
			end

	return @aux
end
GO
```

# [dbo].[vista]

```sql
/****** Object:  View [dbo].[vista]    Script Date: 20/06/2023 20:01:21 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create view [dbo].[vista] as (
select top 1 hotel.nombre as  hotel_mayor_habitacion , COUNT(habitacion.id_habitacion) as Cantidad_habitacion
from hotel inner join habitacion on hotel.id_hotel = habitacion.id_hotel
group by hotel.nombre
order by count (habitacion.id_hotel) desc
);
GO
```
